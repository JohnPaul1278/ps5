Just to get an idea for how the tool works, let's adjust the average hourly pay for a worker with an advanced degree in the year 2000. Will it match the average hourly pay for workers with advanced degrees today? Will it be higher?

```python
cpi.inflate(45.99, 2000)
```

Looks like the hourly rate of forty five ninety nine is worth seventy two dollars and thirty seven cents in today's money.
That's significantly lower than the fifty three dollar and twenty two cent hourly rate that the average worker with an advanced degree earns today. The value of an hour's work has essentially shrunk by nineteen dollars!

The value of an advanced degree is measurably smaller today than it once was.

Now let's adjust the very first minimum wage from 1938 to today's standards. The first minimum wage was only $0.25! How much would that be worth today?


```python
cpi.inflate(0.25, 1938)
```

The first minimum wage implemented in the United States would be worth about 4.80 in today's money. Our current minimum wage, which was set in 2009, is 7.25. Our economy seems to have improved at least a little in the last 85 years.
